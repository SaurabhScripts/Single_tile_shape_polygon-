```python
import pandoc
import os
 
def convert_docx_to_pdf(doc_path, output_folder):
    # Define the output PDF path
    output_pdf_path = os.path.join(output_folder, os.path.splitext(os.path.basename(doc_path))[0] + '.pdf')
    
    # Convert DOCX to PDF using pypandoc
    try:
        pypandoc.convert_file(doc_path, 'pdf', outputfile=output_pdf_path, extra_args=['-V', 'geometry:margin=1in'])
        print(f"Converted {doc_path} to {output_pdf_path}")
    except RuntimeError as e:
        print(f"Failed to convert {doc_path}. Error: {e}")
 
def batch_convert_docx_to_pdf(directory, output_folder):
    # Check if the output folder exists, create if not
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
 
    # Iterate through all files in the directory
    for filename in os.listdir(directory):
        if filename.endswith(".docx"):
            doc_path = os.path.join(directory, filename)
            convert_docx_to_pdf(doc_path, output_folder)
 
# Example usage
source_directory = r'C:\Users\Saurabh.meena\Downloads\2022 Orders'  # Update this path
output_directory =  r'C:\Users\Saurabh.meena\Downloads\new1'  # Update this path
batch_convert_docx_to_pdf(source_directory, output_directory)
```

```python
# [4:26 PM] Saurabh Meena
import os
import pandas as pd
 
def get_subdirectories(path):
    """Returns a list of subdirectory names in the given directory."""
    return [d for d in os.listdir(path) if os.path.isdir(os.path.join(path, d))]
 
def compare_folders(base_path, main_folder, comparison_folders):
    """Compare directories in the main_folder against each of the comparison_folders."""
    main_dir_path = os.path.join(base_path, main_folder)
    main_dirs = set(get_subdirectories(main_dir_path))
    
    results = {}
    
    for folder in comparison_folders:
        comp_dir_path = os.path.join(base_path, folder)
        comp_dirs = set(get_subdirectories(comp_dir_path))
        unique_dirs = main_dirs - comp_dirs
        results[folder] = list(unique_dirs)
    
    return results
 
def main(base_path):
    # Folders to compare
    main_folder = 'Other'
    comparison_folders = ['DD', 'Receipt', 'Loss Assessment']
    
    # Perform comparisons
    comparison_results = compare_folders(base_path, main_folder, comparison_folders)
    
    # Create DataFrame for each comparison and save to Excel
    writer = pd.ExcelWriter(os.path.join(base_path, 'folder_comparisons.xlsx'), engine='xlsxwriter')
    
    for folder, unique_dirs in comparison_results.items():
        df = pd.DataFrame(unique_dirs, columns=[f'Unique in {main_folder} not in {folder}'])
        df.to_excel(writer, sheet_name=f'Not in {folder}', index=False)
    
    writer.save()
    print(f"Excel file has been created and saved at: {os.path.join(base_path, 'folder_comparisons.xlsx')}")
 
if __name__ == "__main__":
    base_path = r'C:\Users\Saurabh.meena\Downloads\PKG_1_Tree_Files_input\PKG_1'  # Update this path to your base directory
    main(base_path)
```

```python
# [4:02 PM] Saurabh Meena
# import os
# from PIL import Image
# import pytesseract
 
# Set the path to the tesseract executable if it's not in your PATH
# pytesseract.pytesseract.tesseract_cmd = r'<full_path_to_your_tesseract_executable>'
 
def correct_orientation(image_path):
    """Correct the orientation of the given image for better OCR results."""
    image = Image.open(image_path)
    
    # Use pytesseract to detect orientation
    osd = pytesseract.image_to_osd(image)
    rotate = 0
    
    # Extract the rotation angle from the OSD output
    for line in osd.split('\n'):
        if 'Rotate' in line:
            rotate = int(line.split(':')[1].strip())
    
    # Rotate the image if necessary
    if rotate != 0:
        image = image.rotate(-rotate, expand=True)  # Negative for PIL rotation
    
    return image
 
def apply_ocr(image):
    """Apply OCR on the given image and return the extracted text."""
    text = pytesseract.image_to_string(image, lang='eng')  # Specify the language as needed
    return text
 
def process_images_in_folder(folder_path):
    """Process all images in the given folder."""
    for filename in os.listdir(folder_path):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(folder_path, filename)
            print(f'Processing {filename}...')
            corrected_image = correct_orientation(image_path)
            text = apply_ocr(corrected_image)
            print(text)
 
# Example usage: Adjust 'folder_path' to your images directory
folder_path = r"C:\Users\Saurabh.meena\Downloads\Testing_DD"
process_images_in_folder(folder_path)
```

```python
# [1:50 PM] Saurabh Meena
import os
import shutil
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
from tqdm import tqdm
import pandas as pd
 
# Define the path to the directory containing PDFs
path_of_dir_1 = r"C:\Users\Saurabh.meena\Downloads\PKG_1_Tree_Files_input\PKG_1"
 
# Define directories for categorized images
categories = ['Loss Assessment', 'Receipt', 'DD', 'Other']
 
# Ensure category directories exist
for category in categories:
    category_dir = os.path.join(path_of_dir_1, category)
    if not os.path.exists(category_dir):
        os.makedirs(category_dir)
 
# DataFrame to store the log data
log_data = []
 
# Helper functions for categorization
def categorize_text_gujarati(extracted_text):
    Lossr_keywords=["નુકસાન આડારણી","ખાતરી","વુક્સાનની","પાક /વૃક્ષ / જમીન"]
    # if "ખાતરી" in extracted_text or "નુકસાન આડારણી" in extracted_text or "વુક્સાનની" in extracted_text or "પાક /વૃક્ષ / જમીન" in extracted_text:
    if any(keyword in extracted_text for keyword in Lossr_keywords):
        return 'Loss Assessment'
    # Multiple Gujarati keywords for "Receipt"
    receipt_keywords = ["જમીત માલિકની","વહોંચ","બાંહેધરી ઘત્ર","સટરી્ગીગ","પહોંચ","જમીન માલિકનીિ","ડીડી" ,"જમીન માલિકની","કળી તંબર","સ્યીનગીંગ","સ્ટરીન્ગીંગ","પહાચ","પણવ","સટરીનગીગ","લોકેશન નંબર","લોકેશનનંબર","સ્ટરીનગીંગ"]
    if any(keyword in extracted_text for keyword in receipt_keywords):
        return 'Receipt'
    return None
 
def categorize_text_english(extracted_text):
    phrases = ["suing Branch","PAYABLE AT","YY YyY","from the date of issue","Not over Rs","purchaser Name","uing Branch","NOT NEGOTIABLE","value received","Cash Management Service","On Demand Pay","Valid for Three","for valu reeivod","ssuing Branch","DoMM YY YY","For value received","OnDemand Pay","DDMN YY YyY","from the date of Issue","cashmanagementservices", "orderof", "escrow", "paymentreference", "dommyy", "scrow", "cashmanagomontsorvicos","Not over Rs"]
    if any(phrase in extracted_text.lower() for phrase in phrases):
        return 'DD'
    return 'Other'
 
# Iterate over each file in the directory
for folder in tqdm(os.listdir(path_of_dir_1)):
    Dir_of_pdfs = os.path.join(path_of_dir_1, folder)
    if os.path.isdir(Dir_of_pdfs):
        for file in os.listdir(Dir_of_pdfs):
            file_path = os.path.join(Dir_of_pdfs, file)
            if file_path.lower().endswith(".pdf"):
                doc = fitz.open(file_path)
                for i, page in enumerate(doc):
                    # Convert the PDF page to an image
                    pix = page.get_pixmap()
                    image_name = f'{file}_{i+1}.png'
                    image_path = os.path.join(Dir_of_pdfs, image_name)
                    pix.save(image_path)
 
                    # Perform Gujarati OCR
                    extracted_text_guj = pytesseract.image_to_string(Image.open(image_path), lang='guj')
                    category = categorize_text_gujarati(extracted_text_guj)
 
                    # If not categorized by Gujarati OCR, use English OCR
                    if not category:
                        extracted_text_eng = pytesseract.image_to_string(Image.open(image_path), lang='eng')
                        category = categorize_text_english(extracted_text_eng)
                    else:
                        extracted_text_eng = ""
 
                    # Build the directory path including category and PDF name
                    pdf_name = os.path.splitext(file)[0]
        # category_path = os.path.join(path_of_dir_1, folder, pdf_name)
                    category_path = os.path.join(path_of_dir_1,category, folder)
 
                    # Ensure this directory exists
                    if not os.path.exists(category_path):
                        os.makedirs(category_path)
 
                    # Move the image to the corresponding folder
                    final_image_path = os.path.join(category_path, f'{pdf_name}_{i+1}.png')
                    shutil.move(image_path, final_image_path)
 
                    # Log data for Excel
                    log_data.append({
                        'Folder Name': folder,
                        'PDF Name': file,
                        'Image Name': image_name,
                        'Extracted Text': extracted_text_guj if category != 'Other' and category != 'DD' else extracted_text_eng,
                        'Category': category
                    })
 
                doc.close()
 
# Convert log data to DataFrame
df = pd.DataFrame(log_data)
 
# Save the DataFrame to an Excel file
excel_path = os.path.join(path_of_dir_1, 'extraction_log.xlsx')
df.to_excel(excel_path, index=False)
 
print("Processing and logging complete.")
```

```python
import fitz  # PyMuPDF
import os
 
def convert_pdf_to_images(pdf_path, output_folder):
    # Open the provided PDF file
    document = fitz.open(pdf_path)
    
    # Iterate through each page
    for page_number in range(len(document)):
        page = document.load_page(page_number)  # load page
        pix = page.get_pixmap()  # render page to an image
        output_path = os.path.join(output_folder, f"page_{page_number + 1}.png")
        pix.save(output_path)  # save image to file
 
# Example usage
convert_pdf_to_images(r"C:/Users/Saurabh.meena/Downloads/Rewari/Rewari/2024-EROLLGEN-S07-74-FinalRoll-Revision1-HIN-9-WI.pdf", r'C:/Users/Saurabh.meena/Downloads/Rewari')
```
